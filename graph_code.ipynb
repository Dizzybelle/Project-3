```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
```

```python
life_exp_data = Path('data\life_exp.csv')
pop_data = Path('data\cities_lat_long_clean_rename.csv')
```

```python
life_exp_graph = pd.read_csv(life_exp_data)
population_graph = pd.read_csv(pop_data)
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(life_exp_graph['Life_expectancy']))
tick_locations = [value for value in x_axis]
```

```python
plt.figure(figsize=(20,4))
plt.title('Life Expectancy by Prefecture')
plt.plot(x_axis, life_exp_graph['Life_expectancy'], color='g', alpha=0.5)
plt.xticks(tick_locations, life_exp_graph['Prefecture'], rotation="vertical")
plt.savefig('life_exp_graph2.png')
plt.show()
plt.tight_layout()
```

```python
graph_data_need = population_graph[['admin_name', 'population']]
graph_data = graph_data_need.groupby(['admin_name'])
pop_graph = graph_data.sum()
pop_graph
```

```python
pop_graph.plot(kind='bar', color=['red'], figsize=(12,4), title='Population by Prefecture', xlabel='Prefecture')
plt.savefig('population_graph2.png')
plt.show()
plt.tight_layout()
```
